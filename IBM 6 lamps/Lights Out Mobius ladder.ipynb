```python
import itertools
from qiskit.quantum_info import SparsePauliOp

import itertools
from typing import List
from qiskit.quantum_info import SparsePauliOp

def projector_on_subset(
    bitstring_msb_to_lsb: str,
    qubits_msb_to_lsb: List[int],
    num_qubits_total: int
) -> SparsePauliOp:
    """
    Returns _bitstring  I_rest as a SparsePauliOp on num_qubits_total qubits.
    """
    m = len(bitstring_msb_to_lsb)
    assert m == len(qubits_msb_to_lsb)

    term_coeffs = {}
    for mask in itertools.product([0, 1], repeat=m):
        label = ["I"] * num_qubits_total
        sign = 1.0
        for j, use_z in enumerate(mask):
            if use_z:
                q = qubits_msb_to_lsb[j]
                # Qiskit Pauli labels: right-most char is qubit 0 :contentReference[oaicite:1]{index=1}
                label[num_qubits_total - 1 - q] = "Z"
                if bitstring_msb_to_lsb[j] == "1":
                    sign *= -1.0

        coeff = sign / (2 ** m)
        key = "".join(label)
        term_coeffs[key] = term_coeffs.get(key, 0.0) + coeff

    return SparsePauliOp.from_list(list(term_coeffs.items()))


def success_projector(good_bitstrings, num_qubits_total, qubits_msb_to_lsb):
    op = None
    for b in good_bitstrings:
        pb = projector_on_subset(b, qubits_msb_to_lsb, num_qubits_total)
        op = pb if op is None else (op + pb)
    return op
```
